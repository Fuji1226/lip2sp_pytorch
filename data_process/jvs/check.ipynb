```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_path = '/home/minami/dataset/jvs_ver1/gender_f0range.txt'
df = pd.read_csv(data_path)
df['data_split'] = ['train' for i in range(90)] + ['val' for i in range(10)]
df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
sns.scatterplot(df, x='minf0[Hz]', y='maxf0[Hz]', hue='Male_or_Female', ax=ax[0])
sns.scatterplot(df, x='minf0[Hz]', y='maxf0[Hz]', hue='data_split', ax=ax[1])
ax[0].grid()
ax[1].grid()
plt.show()
```
